```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
#on va chercher la page suivante :
page = requests.get("https://store.nike.com/fr/fr_fr/pw/homme-running-chaussures/7puZ8yzZoi3")
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Pour nous assurer que le site est accessible, nous pouvons
# nous assurer que nous obtenons une réponse 200 OK pour indiquer
# que la page est bien présente:
page.status_code
```

```python
# vérifier que nous avons bien accédé à la bonne page:
print(page.headers)
```

```python
chaussure_class_data = soup.find(class_='exp-gridwall-standard')
```

```python
print(chaussure_class_data.prettify())
```

```python
data = chaussure_class_data.find_all(class_='grid-item nikeid')
```

```python
#on affiche une entrée
print(data[0].prettify())
```

```python
#on identifie et récupére le titre
product_name = data[0].find('p', class_='product-display-name').get_text()
product_name
```

```python
#on récupére la description
product_desc = data[0].find('p', class_='product-subtitle').get_text()
product_desc
```

```python
#on récupére le prix
price = data[0].find('span', class_='local nsg-font-family--base').get_text()#on sélection 
price
```

```python
#on récupere la donnée nombre de couleurs
color = data[0].find('div', class_='number-of-colors').get_text()
color
```

```python
#on récupére l'image
image = data[0].find('img')
image
```

```python
#on enleve la balise src
image_display_link = image['src']
```

```python
image_display_link
```
